```julia
using Colors
using ColorVectorSpace
```

```julia
function cdemo((c1, c2), x)
  return c1 + (c2 - c1)*x
end

colours = [
    (
        RGB(0, 0, 0),
        RGB(1, 1, 1)
    ),
    (
        RGB(1, 0, 0),
        RGB(1, 0, 1)
    ),
    (
        RGB(0, 1, 1),
        RGB(0.2, 1, 0.5)
    )
]
display([cdemo(colours[1], x) for x in 0:0.001:1])
display([cdemo(colours[2], x) for x in 0:0.001:1])
display([cdemo(colours[3], x) for x in 0:0.001:1])
```
