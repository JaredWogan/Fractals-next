```julia
savedir = "C:\\Users\\Jared\\OneDrive\\Github\\Fractals-next\\Julia\\Images";
```

```julia
function func(z, f_cache, iteration, param)
    fp = f_cache[iteration - 1]
    return param * (1 - fp) * (1 + fp) * fp
end

gen_fractal(
    func=func,
    param=complex(1.18, 0.376),
    max_iters=20,
    div=1000,
    w=1024,
    h=1024,
    xrange=1.5,
    yrange=1.5,
    xcenter=0,
    ycenter=0,
    zoom=1,
    coord_z0=true,
    cmap="Spectral",
    savedir=savedir,
    invert_cmap=false
)
```
